```python
class SinglyLinkedList:
    class Node: 
        def __init__(self, data):
            self.data = data
            self.next = None    # references next node
            
    def __init__(self):
        self.head = None        # references head node - head is base of entire list *Do Not Delete Head Node*
                                # tail node will always reference none - end of list
    def append(self, value):
        new_node = self.Node(value)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            
    def__str__(self):
        out = "["
        current = self.head
        if current:
            out += "%s" % current.data
            current = current.next
            while current:
                out += ", %s" % current.data
                current = current.next
        out += "]"
        return out
    
    def index(self, value):
        if self.head:
            counter = 0
            current = self.head
            while current.next and current.data != value:
                current = current.next
                counter += 1
            if current.data == value:
                return counter
        raise ValuError("Value not found")
       
    
    def insert(self, index, value):
        new_node = self.Node(value)
        if not self.head:
            self.head = new_node
        else:
            if index == 0, 
                of_head = self.head
                self.head = new_node
                new_node.next = old_head
            else:
                current = self.head
                prev = None
                counter = 0
                while current.next and index != counter:
                    prev = current
                    current = current.next
                    counter += 1
                if index == counter:
                    prev.next = new_node
                    new_node.next = current
                else: 
                    current.next = new_node
                    
        
        
    def remove(self, value):
        # remove the first occurence value (and node)
        # raise ValueError if not present
        pass
            
```
